# Matplotlib 

When diving into data visualization in Python, you'll quickly encounter two powerhouse libraries: **Matplotlib**. Let's break down what makes each one special and when you might want to use them.

## Matplotlib: The Swiss Army Knife of Python Plotting

Think of Matplotlib as your trusty Swiss Army knife for data visualization. It's been around since 2003 and has become the backbone of Python plotting.

### What Makes Matplotlib Special?
Matplotlib gives you incredible control over every tiny detail of your plots. Want to adjust the exact position of a tick mark? Change the font size of just one label? Customize the line width of your axes? Matplotlib has got you covered.

- **Fine-grained control**: You can tweak literally everything - from axis positions to color gradients
- **Publication-ready**: Creates crisp, professional figures perfect for research papers and reports  
- **Plays nice with friends**: Works seamlessly with NumPy arrays and Pandas DataFrames
- **Flexible output**: Save your plots as PNG, PDF, SVG, or display them interactively

### When Should You Reach for Matplotlib?
- You're doing academic research and need pixel-perfect figures
- You want to create a completely custom plot that doesn't fit standard templates
- You're exploring data and need quick, simple plots
- You're building a dashboard or application where you need precise control over the visual elements


# Graph Types
## Matplotlib Graph Types
### 1.Line Plot

Description: Represents data trends over time or continuous variables.

Use Case: Plotting stock prices over time.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, marker='o')
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### 2.Bar Chart

Description: Used to compare categories.

Use Case: Comparing sales across products.

```python
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

plt.bar(categories, values, color='skyblue')
plt.title("Bar Chart Example")
plt.show()
```

### 3.Scatter Plot

Description: Shows relationships between two variables.

Use Case: Relationship between height and weight.

```python
x = [5, 7, 8, 7, 6, 9, 5, 6]
y = [99, 86, 87, 88, 100, 86, 103, 87]

plt.scatter(x, y, color='red')
plt.title("Scatter Plot Example")
plt.show()
```

### 4.Histogram

Description: Distribution of a dataset.

Use Case: Distribution of exam scores.

```python
import numpy as np

data = np.random.randn(50)
plt.hist(data, bins=50, color='green', alpha=0.7)
plt.title("Histogram Example")
plt.show()
```

## Seaborn: The Elegant Statistician's Choice

If Matplotlib is a Swiss Army knife, then Seaborn is like having a professional chef's knife - more specialized, but incredibly effective at what it does. Built on top of Matplotlib, Seaborn focuses on making statistical visualizations both beautiful and effortless.

### What Makes Seaborn Shine?
Seaborn takes the philosophy of "make complex things simple." It's designed with statisticians and data scientists in mind, offering high-level functions that create gorgeous plots with minimal code.

- **Statistical focus**: Built-in functions for correlation matrices, distribution plots, and regression visualizations
- **DataFrame integration**: Works directly with Pandas DataFrames - just pass your column names!
- **Aesthetic by default**: Comes with beautiful themes and color palettes that make your plots look professional instantly
- **Less code, more insight**: Create complex statistical visualizations in just a few lines

### When Should You Choose Seaborn?
- You're exploring a dataset and want to quickly understand relationships between variables
- You need statistical plots like violin plots, box plots, or heatmaps
- You want your plots to look great without spending time on styling
- You're working with structured data in Pandas and want seamless integration
- You're doing machine learning and need to visualize feature relationships


## B.Seaborn Graph Types
### 1. Line Plot

Description: Automatically handles Pandas data and adds confidence intervals.

Use Case: Visualizing trends in time-series data.

```python
import seaborn as sns
import pandas as pd

data = pd.DataFrame({
    "x": [1, 2, 3, 4, 5],
    "y": [2, 4, 6, 8, 10]
})

sns.lineplot(x="x", y="y", data=data, marker="o")
```

### 2. Bar Plot

Description: Shows mean values and confidence intervals for categorical data.

Use Case: Comparing average test scores by gender.

```python
tips = sns.load_dataset("tips")
sns.barplot(x="day", y="total_bill", data=tips)
```

### 4. Heatmap

Description: Visualizes data as color-coded matrices.

Use Case: Correlation matrix in machine learning datasets.

```python
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
```

### Comparison: Matplotlib vs Seaborn

- **Matplotlib**  
  - Low-level library with detailed control.  
  - Supports all basic plots (line, bar, scatter, histogram, pie).  
  - Highly customizable but requires more code.  
  - Default style is plain; needs manual styling.  

- **Seaborn**  
  - High-level library built on top of Matplotlib.  
  - Best for statistical and advanced plots (heatmap, violin, pairplot).  
  - Very simple syntax with attractive default themes.  
  - Limited customization compared to Matplotlib.
